```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Count plot
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Bachelor', hue='output', palette='Blues')
plt.title('Count Plot of Bachelor Degree with Job Success')
plt.xlabel('Bachelor Degree')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Graduated', 'Graduated'])
# Annotate each bar with its count
for p in ax.patches:
    if p.get_height() != 0:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Get legend handles and labels
handles, labels = ax.get_legend_handles_labels()

# Define custom legend labels
custom_labels = {'0': 'Not Offered', '1': 'Offered'}

# Update legend labels
for i, label in enumerate(labels):
    if label in custom_labels:
        labels[i] = custom_labels[label]

# Set custom legend labels
plt.legend(handles, labels, title='Job Success')

plt.show()

```

```python
import plotly.figure_factory as ff
from plotly.subplots import make_subplots

# Confusion Matrix for Ann Model Model
ann_conf_matrix = [[10858, 803],
                  [31 , 5683]]

# Confusion Matrix for XGBoost Model
gbm_conf_matrix = [[10852,809],
                   [23, 5691]]

# Class labels
class_labels = ['Positive', 'Negative']

# Create subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=("Artificial Neural Network (ANN)", "Gradient Boosting Machine"))

# Plotting Random Forest Confusion Matrix
ann_fig = ff.create_annotated_heatmap(z=ann_conf_matrix,
                                     x=class_labels,
                                     y=class_labels,
                                     colorscale='YlGnBu',
                                     showscale=False)  # Hide color scale

# Plotting XGBoost Confusion Matrix
gbm_fig = ff.create_annotated_heatmap(z=gbm_conf_matrix,
                                      x=class_labels,
                                      y=class_labels,
                                      colorscale='YlGnBu',
                                      showscale=False)  # Hide color scale

# Add subplots
fig.add_trace(ann_fig.data[0], row=1, col=1)

# Add values inside cells for Random Forest Confusion Matrix
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        font_color = 'white' if ann_conf_matrix[i][j] == 10858 else 'black'
        fig.add_annotation(text=str(ann_conf_matrix[i][j]),
                           x=class_labels[j], y=class_labels[i],
                           showarrow=False, font=dict(color=font_color), row=1, col=1)

# Add XGBoost Confusion Matrix
fig.add_trace(gbm_fig.data[0], row=1, col=2)

# Add values inside cells for gbm Confusion Matrix
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        font_color = 'white' if gbm_conf_matrix[i][j] == 10852 else 'black'
        fig.add_annotation(text=str(gbm_conf_matrix[i][j]),
                           x=class_labels[j], y=class_labels[i],
                           showarrow=False, font=dict(color=font_color), row=1, col=2)

# Update layout
fig.update_layout(title_text="Confusion Matrix Comparison", title_x=0.5,
                  height=500, width=800,
                  margin=dict(l=0, r=5, t=50, b=50),  # Set margins (l=left, r=right, t=top, b=bottom)
                  plot_bgcolor='rgba(0,0,0,0)')  # Set plot background color to transparent

fig.show()
```
